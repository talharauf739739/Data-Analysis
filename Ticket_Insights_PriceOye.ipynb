```python
import pandas as pd

# Assuming tickets and order are already defined pandas DataFrames

# Step 1: Get the unique subjects from the tickets dataset
ticket_subjects = tickets['subject'].unique()

# Step 2: Filter the orders dataset based on the order_id matching the subjects from tickets
back_orders = order[order['order_id'].isin(ticket_subjects)]

# Step 3: Extract the id column from the tickets dataset and rename it to 'ticket_id'
ticket_ids = tickets[tickets['subject'].isin(ticket_subjects)][['id', 'subject']]
ticket_ids = ticket_ids.rename(columns={'id': 'ticket_id', 'subject': 'order_id'})

# Step 4: Merge the extracted ticket ids with the filtered orders dataset
back_orders = pd.merge(back_orders, ticket_ids, on='order_id', how='left')

# Display the back_orders dataset
back_orders
```

```python
back_orders.to_excel('Tickets_of_June_and_July_with_categories.xlsx', index  = False)
```
